```python
nodes = pandas.read_csv('ptgm_nodes.csv', index_col='nid').astype(object)
links = pandas.read_csv('ptgm_links.csv', index_col='lid').astype(object)
```

```python
def basemap_pakistan(resolution='l', proj='merc', fix_aspect=False, **kwargs):
    m = mpl_toolkits.basemap.Basemap(llcrnrlon=58, llcrnrlat=23, urcrnrlon=80, urcrnrlat=39,
                                     resolution=resolution,
                                     fix_aspect=fix_aspect, **kwargs)
    m.projection = proj
    m.drawcoastlines()
    m.drawcountries()
    return m
```

```python
figsize(9, 9)
m = basemap_pakistan()

links500 = links[links.voltage == 500000]
plot([], [], color='red', label='500 kV')
for _, link in links500.iterrows():
    n1 = nodes.loc[link.nid1]
    n2 = nodes.loc[link.nid2]
    m.plot([n1.lon, n2.lon], [n1.lat, n2.lat], color='red')

links220 = links[links.voltage == 220000]
plot([], [], color='green', label='220 kV')
for _, link in links220.iterrows():
    n1 = nodes.loc[link.nid1]
    n2 = nodes.loc[link.nid2]
    m.plot([n1.lon, n2.lon], [n1.lat, n2.lat], color='green')

m.scatter(nodes.lon, nodes.lat, latlon=True, color='k', marker='.')
legend(loc='best')
savefig('ptgm_plot.pdf', bb_inches='tight');
```
